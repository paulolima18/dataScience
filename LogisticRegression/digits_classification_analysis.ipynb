# Digit Classification Analysis

The objective of this project is to make use of logistic regression models to predict a given digit from the dataset.

This dataset is available at UCI ML housing dataset.
https://archive.ics.uci.edu/ml/datasets/Optical+Recognition+of+Handwritten+Digits


#### The data contains the following columns
* [0..9]
    - 10 columns from 0 to 9.

## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Get Data

```python
from sklearn.datasets import load_digits
data = load_digits()
```

## Exploratory Data Analysis

```python
plt.figure(1, figsize=(5, 5))
plt.imshow(data.images[0], cmap = plt.cm.Greens_r, interpolation = 'nearest')
plt.show()
```

```python
plt.figure(figsize=(20,9))
for index in range(10):
    plt.subplot(2, 5, index + 1)
    plt.imshow(data.images[index], cmap = plt.cm.RdBu, interpolation = 'hanning')
    plt.title('Training: {0}\n'.format(index), fontsize = 20)
```

## Training and Testing Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size = 0.2, random_state = 101)
```

## Training the Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
logm = LogisticRegression(max_iter = 3000)
logm.fit(X_train, y_train)
```

```python
plt.figure(figsize=(20,9))
for index in range(10):
    plt.subplot(2, 5, index + 1)
    plt.imshow(logm.coef_[index].reshape(8,8), cmap = plt.cm.RdBu, interpolation = 'hanning')
    plt.title('Trained: {0}\n'.format(index), fontsize = 20)
```

## Predicting Test Data

```python
predictions = logm.predict(X_test)
```

## Evaluating the Model

```python
from sklearn import metrics
```

```python
cm = metrics.confusion_matrix(y_test, predictions)
score = metrics.accuracy_score(y_test, predictions)
```

```python
all_sample_title = 'Accuracy Score: {0}'.format(score)
plt.figure(figsize=(10,10))
sns.heatmap(cm, annot=True, linewidths=.5, square = True, cmap = plt.cm.Greens_r)
plt.xlabel('Predicted label')
plt.ylabel('Actual label')
plt.title(all_sample_title)
```

## Conclusion

Evaluating the model it is possible to see that the model is underfitting.
The used dataset is small, to solve this problem, a solution could be to feed more information to the model.
